```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("cancerdata.csv"))
df['diagnosis'] = df['diagnosis'].replace(['B'], 0)
df['diagnosis'] = df['diagnosis'].replace(['M'], 1)
df['diagnosis'] = pd.to_numeric(df['diagnosis'])
train = df.sample(frac=0.8, random_state=25)
test = df.drop(train.index)
df['diagnosis']
```

```python
for (index, column) in enumerate(train):
    if(index > 1):
        print(index,column)
```

```python
def normalize(X):
    result = np.zeros(X.shape)
    for (index, colname) in enumerate(train):
        if(index > 1 and index < 32):
            result[:,index-2] = (train[colname].values-train[colname].values.min())/(train[colname].values.max()-train[colname].values.min())
    return result
```

```python
X = np.zeros((455,30))
X = normalize(X)
X
```
